Экспертная система предлогает помощь в вопросе приманок для ловли рыбы, для этого в экспертной системе существует 2 фактора:

TimeFact - Факт, отображающий время дня, когда собираются рыбачить.

WaterTemperatureFact - Факт, отображающий температуру воды на момент рыбалки.

Таким образом, отталкиваясь от 2 основных факторов создается 3 факт FishFact - Факт, отображающий вид рыбы и приманку для ее ловли, отображающий какуй рыбу и на какую приманку лучше всего ловить в заданных условиях.

Сама система сотоит из 6 правил для каждого набора комбинаций, время дня отображается в формате - {Утро, День, Вечер}, температура - промежуточное значение между 10 и 30 градусами Цельсия.

```python
import random 
from experta import *

class TimeFact(Fact):
    """Факт, отображающий время дня."""
    pass

class WaterTemperatureFact(Fact):
    """Факт, отображающий температуру воды."""
    pass

class FishFact(Fact):
    """Факт, отображающий вид рыбы и приманку для ее ловли."""
    pass

class FishingRecommendation(KnowledgeEngine):
    """Экспертная система для предоставления рекомендаций по рыбалке.""" 
    @Rule(AND(TimeFact(TimeOfDay='утро')),
    (WaterTemperatureFact(WaterTemperature=P(lambda WaterTemperature: WaterTemperature >= 10) 
                               & P(lambda WaterTemperature: WaterTemperature <= 20))))   
    def rule_1(self):
        """Рекомендация для утренней рыбалки в холодную погоду."""
        self.declare(FishFact(name="форель", bait="червь"))
        
        print(self.facts)
        
        print(f"Если сейчас {self.facts[1]['TimeOfDay']} и температура воды {self.facts[2]['WaterTemperature']} градусов, попробуйте порыбачить на {self.facts[3]['name']} с помощью  {self.facts[3]['bait']}.")

    @Rule(AND(TimeFact(TimeOfDay='утро')),
    (WaterTemperatureFact(WaterTemperature=P(lambda WaterTemperature: WaterTemperature >= 21) 
                               & P(lambda WaterTemperature: WaterTemperature <= 30))))  
    def rule_2(self):
        """Рекомендация для утренней рыбалки в теплую погоду."""
        self.declare(FishFact(name="окунь", bait="воблер"))
        print(self.facts)
        
        print(f"Если сейчас {self.facts[1]['TimeOfDay']} и температура воды {self.facts[2]['WaterTemperature']} градусов, попробуйте порыбачить на {self.facts[3]['name']} с помощью {self.facts[3]['bait']}.")

    @Rule(AND(TimeFact(TimeOfDay='день')),
    (WaterTemperatureFact(WaterTemperature=P(lambda WaterTemperature: WaterTemperature >= 10) 
                               & P(lambda WaterTemperature: WaterTemperature <= 20)))) 
    def rule_3(self):
        """Рекомендация для дневной рыбалки в холодную погоду."""
        self.declare(FishFact(name="щука", bait="воблер"))
        print(self.facts)
        
        print(f"Если сейчас {self.facts[1]['TimeOfDay']} и температура воды {self.facts[2]['WaterTemperature']} градусов, попробуйте порыбачить на {self.facts[3]['name']} с помощью {self.facts[3]['bait']}.")

    @Rule(AND(TimeFact(TimeOfDay='день')),
    (WaterTemperatureFact(WaterTemperature=P(lambda WaterTemperature: WaterTemperature >= 21) 
                               & P(lambda WaterTemperature: WaterTemperature <= 30)))) 
    def rule_4(self):
        """Рекомендация для дневной рыбалки в теплую погоду."""
        self.declare(FishFact(name="сом", bait="червь"))
        print(self.facts)
        
        print(f"Если сейчас {self.facts[1]['TimeOfDay']} и температура воды {self.facts[2]['WaterTemperature']} градусов, попробуйте порыбачить на {self.facts[3]['name']} с помощью {self.facts[3]['bait']}.")

    @Rule(AND(TimeFact(TimeOfDay='вечер')),
    (WaterTemperatureFact(WaterTemperature=P(lambda WaterTemperature: WaterTemperature >= 10) 
                               & P(lambda WaterTemperature: WaterTemperature <= 20)))) 
    def rule_5(self):
        """Рекомендация для вечерней рыбалки в холодную погоду."""
        self.declare(FishFact(name="судак", bait="мормышка"))
        print(self.facts)
        
        print(f"Если сейчас {self.facts[1]['TimeOfDay']} и температура воды {self.facts[2]['WaterTemperature']} градусов, попробуйте порыбачить на {self.facts[3]['name']} с помощью {self.facts[3]['bait']}.")

    @Rule(AND(TimeFact(TimeOfDay='вечер')),
    (WaterTemperatureFact(WaterTemperature=P(lambda WaterTemperature: WaterTemperature >= 21) 
                               & P(lambda WaterTemperature: WaterTemperature <= 30)))) 
    def rule_6(self):
        """Рекомендация для вечерней рыбалки в теплую погоду."""
        self.declare(FishFact(name="уклейка", bait="муха"))
        print(self.facts)
        
        print(f"Если сейчас {self.facts[1]['TimeOfDay']} и температура воды {self.facts[2]['WaterTemperature']} градусов, попробуйте порыбачить на {self.facts[3]['name']} с помощью {self.facts[3]['bait']}.")

engine = FishingRecommendation()

engine.reset()
time=random.choice(['утро','день','вечер'])
temp=random.randint(10, 30)
engine.declare(TimeFact(TimeOfDay = time))

engine.declare(WaterTemperatureFact(WaterTemperature = 21))
print(f"-------1-ый вызов (время - {time}, температура - {temp} градусов)----------")
engine.run()
#--------------------------------------------------------------------------------#
engine.reset()
time=random.choice(['утро','день','вечер'])
temp=random.randint(10, 30)
engine.declare(TimeFact(TimeOfDay = time))

engine.declare(WaterTemperatureFact(WaterTemperature = 21))
print(f"-------2-ой вызов (время - {time}, температура - {temp} градусов)----------")
engine.run()
```
