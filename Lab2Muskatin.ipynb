# 1 Иерархическая кластеризация (Цветы Ириса)

Подключим исходные данные

```python
data = pd.read_csv('IrisCategorical.csv', index_col='Id')
data
```

Выполним визуализацию данных по двум атрибутам. Сначала возьмем атрибут ирисов - Sepal, затем рассмотрим атрибут Petal.

Исходя из графической интерпретации исходных данных нет явно выделяющихся групп. 
Построим дендрограмму, для дальнейшего анализа.

## Sepal

```python
new_data=data.drop(columns=['Species', 'PetalLengthCm', 'PetalWidthCm'], axis=1) # рассмотрим дендрограмму для атрибута Sepal
```

```python
sns.clustermap(new_data, cmap='mako')
```

Далее выполним иерархическую кластеризацию используя метод single (MIN).

```python
z = linkage(new_data,'single') # z-матрица содержащая информацию об объединении "точек" данных
print("z has %d rows"%(z.shape[0]))
```

Графически изобразим данные полученные в коде выше. Синяя линия в полученном графике показывает рекомендации для возможного среза.

```python
plt.figure(figsize=(25,10))
plt. title('Дендаграмма Иерархической кластеризации')
plt.xlabel('Iris')
plt.ylabel('distance')
dendrogram(z,
    leaf_rotation=90.,
    leaf_font_size=8.,
)
plt.show()
```

Т.к в задании одним из условий явлется K=5, то выполним "приближение" 5 объединенных кластеров.

```python
plt.title('Дендаграмма Иерархической кластеризации - обрезанная до 5 объединенных кластеров')
 
plt.xlabel('Iris')
plt.ylabel('distance')
dendrogram(
    z,
    truncate_mode='lastp',
    p=5, # количество объединенных кластеров
    show_leaf_counts=False, 
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True, 
)
plt.show()
```

Согласно синей линии мы можем разделить исходные данные на 2,3,4 кластера.

#### Knee/Eblow анализ 
Для решения задачи кластеризации также необходимо вычислим критерии WSS и BSS. Для удобства последующего использования создадим соответствующую функцию.

```python
def KneeEblow(data, max_k):
    wss_values = []
    bss_values = []
    k_values=range(1, max_k)
    
    for k in k_values:
        clustering = fcluster(z, k, criterion='maxclust')
        centroids = [np.mean(data[clustering==c],axis=0) for c in range(1,k+1)]
        cdist(data, centroids, 'euclidean')
        D = cdist(data, centroids, 'euclidean')
        cIdx = np.argmin(D,axis=1)
        d = np.min(D,axis=1)
        avgWithinSS = sum(d)/data.shape[0]
        wss= sum(d**2) 
    
        tss = sum(pdist(data)**2)/data.shape[0]
        bss=tss-wss
        wss_values += [wss]
        bss_values += [bss]
    
    
    fig = plt.figure()
    
    font={'family': 'sans', 'size': 12}
    plt.rc('font', **font)
    
    plt.plot(k_values, wss_values, 'bo-', color='red', label='WSS')
    plt.plot(k_values, bss_values, 'bo-', color='blue', label='BSS')
    plt.grid(True)
    plt.xlabel('Количество кластеров')
    plt.ylabel('BSS & WSS')
    plt.xticks(k_values)
    plt.legend()
    plt.title('Knee/Eblow анализ');  
```

```python
KneeEblow(new_data, 5)
```

Исходя из визуального анализа Knee/Eblow можем предположить что есть небольшие переломы в районе 2 и 3 кластера, но они не четко видны. Проведём дополнительную визуализацию, для более точного определения количества кластеров. 

```python
plt.figure(figsize=(12,12));
for k in range(2,4):
    yp=fcluster(z, k, criterion='maxclust')
    plt.subplot(330 +(k-1))
    plt.title('k='+str(k))
    plt.scatter(data['SepalLengthCm'],data['SepalWidthCm'], c=yp, cmap=plt.get_cmap('tab20b'))
plt.tight_layout()
```

Визуальноый анализ дает основание считать что в случае с анализом по Sepal методом Min получим 1 большой кластер, т.к. при отображении 2 и 3 кластеров, входящие в них элементы больше похожи на "помехи".

# 2 Неирархическая кластеризация k-means (Сегментация рынка)

Подгружаем данные

```python
raw_data = pd.read_csv('MarketSegmentation.csv')
raw_data.head(10)
```

```python
X = raw_data [['Satisfaction', 'Loyalty']]
X = np.array(X) 
```

Представим наши данные графически

```python
plt.figure(figsize=(8,6));
font={'family': 'sans', 'size': 18}
plt.rc('font', **font)
plt.scatter(X[:,0], X[:,1], cmap=plt.get_cmap('Pastel1'));
plt.title("Точки данных")
```

Приступим к выполненнию кластеризации методом K-means, в качестве количества кластеров зададим k=9, как указанно в задании, в качестве метода расчета расстояния между точками будем использовать метод Эвклидова расстояния.

```python
plt.figure(figsize=(12,12));
for i in range(9):
    z=KMeans(n_clusters=(i+2)).fit_predict(X)
    plt.subplot(330 +(i+1))
    plt.title('k='+str(i+2))
    plt.scatter(X[:,0], X[:,1], c=z, cmap=plt.get_cmap('tab20b'))
plt.tight_layout()
```

Визуально могу предположить что подходящее количество кластеров варьируется от 2 до 5, для дальнейшего анализа воспользуемся  методом Knee/Elbow

```python
def KneeEblow(data, max_k):
    k_values=range(1, max_k)
    clusterings=[KMeans(n_clusters=k).fit(data) for k in k_values]
    centroids = [clustering.cluster_centers_ for clustering in clusterings]
    D_k= [cdist(data, cent, 'euclidean') for cent in centroids]
    dist= [np.min(D, axis=1) for D in D_k]
    avgWithinSS = [sum(d)/data.shape[0] for d in dist]
    
    wss= [sum(d**2) for d in dist]
    
    tss = sum(pdist(data)**2)/data.shape[0]
    bss=tss-wss
    
    fig = plt.figure()
    
    font={'family': 'sans', 'size': 12}
    plt.rc('font', **font)
    plt.xticks(k_values)
    plt.plot(k_values, wss, 'bo-', color='red', label='WCSS')
    plt.plot(k_values, bss, 'bo-', color='blue', label='BCSS')
    plt.grid(True)
    plt.ylabel('BSS & WSS')
    plt.xlabel('Количество кластеров')
    plt.legend()
    plt.title('Knee/Eblow анализ');   
```

```python
KneeEblow(X,9)
```

Стоит обратить внимание, что на промежутке 2-4 видны выраженные изменения критерия WCSS и BCSS.
Возьмем для проверки крайний правый локоть/колено k=4.

```python
z = KMeans(n_clusters=4).fit_predict(X)
plt.figure(figsize=(5,5));
font={'family': 'sans', 'size': 18}
plt.rc('font', **font)
plt.scatter(raw_data.iloc[:,1], raw_data.iloc[:,2], c=z, cmap=plt.get_cmap('tab20b'));
```

Из полученного графика можно утверждать что синяя группа клиентов(левый нижний угол) готова покинуть компанию.
